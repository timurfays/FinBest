```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from community import modularity
from networkx.algorithms.community import label_propagation_communities
from sqlalchemy import create_engine
import random
import warnings
warnings.filterwarnings('ignore')

# Подключение к PostgreSQL
engine = create_engine('postgresql://finbest:finbest_password@postgres:5432/finbest')

# Загружаем данные
graph_data = pd.read_sql('SELECT * FROM graph.client_communities', engine)
tx_data = pd.read_sql('SELECT client_id, transaction_id, datetime, amount, currency, merchant, transaction_type, ip_network, category FROM raw.masked_transactions', engine)

print(tx_data.columns)

# Создаем граф
G = nx.Graph()
for _, row in graph_data.iterrows():
    G.add_node(row['client_id'], community_id=row['community_id'], influence_score=row['influence_score'], w_degree=row['w_degree'])

# Добавляем ребра (на основе транзакций P2P и общих хабов)
p2p_edges = tx_data[tx_data['transaction_type'] == 'p2p'][['client_id', 'recipient_id_hash']].dropna()
for _, row in p2p_edges.iterrows():
    G.add_edge(row['client_id'], row['recipient_id_hash'], weight=2.0, layer='p2p')

# Метрики графа
n = G.number_of_nodes()
m = G.number_of_edges()
density = nx.density(G)
avg_deg = round(2*m/n, 2) if n else 0
avg_clust = nx.average_clustering(G, weight='weight')
largest_cc = max(nx.connected_components(G), key=len) if G.nodes() else set()
diameter_lcc = nx.diameter(G.subgraph(largest_cc)) if largest_cc else 0
louvain = nx.get_node_attributes(G, 'community_id')
lou_mod = modularity(louvain, G, weight='weight') if louvain else 0

metrics = pd.Series({
    'nodes': n,
    'edges': m,
    'density': density,
    'avg_weighted_degree': avg_deg,
    'avg_clustering': avg_clust,
    'diameter_LCC': diameter_lcc,
    'modularity_louvain': lou_mod
})
print(metrics.to_markdown())
```

```python
# Круговая диаграмма размеров сообществ
sizes = pd.Series(louvain).value_counts().sort_values(ascending=False)
top10 = sizes.head(10)
fig, ax = plt.subplots(figsize=(7,7))
ax.pie(top10, labels=[f'C{c}' for c in top10.index],
       autopct='%1.1f%%', pctdistance=0.8, startangle=90)
ax.set_title('Top-10 community share (Louvain)')
plt.savefig('/reports/community_sizes.png')
plt.close()
```

```python
# Гистограмма лог-взвешенной степени
deg = pd.Series(dict(G.degree(weight='weight')))
sns.displot(np.log1p(deg), kde=True, height=4, aspect=1.6)
plt.title('Histogram of log-weighted degrees')
plt.xlabel('log1p(weighted degree)')
plt.ylabel('freq')
plt.savefig('/reports/degree_histogram.png')
plt.close()
```

```python
# Функция для визуализации подграфа
def draw_subgraph(comm_attr, comm_name, k=700):
    base_nodes = random.sample(list(G.nodes), min(k, len(G)))
    nbrs = set(base_nodes)
    for n in base_nodes:
        nbrs.update(G[n])
    SG = G.subgraph(nbrs).copy()
    comms = nx.get_node_attributes(G, comm_attr)
    colors = [comms.get(v, -1) for v in SG.nodes()]
    cmap = plt.cm.get_cmap('tab20')
    pos = nx.spring_layout(SG, seed=42, k=0.2)
    plt.figure(figsize=(10,8))
    nx.draw_networkx_nodes(SG, pos, node_size=40, node_color=colors, cmap=cmap, alpha=0.9)
    nx.draw_networkx_edges(SG, pos, width=0.3, alpha=0.4)
    plt.title(f'Sampled graph coloured by {comm_name} communities')
    plt.axis('off')
    plt.savefig(f'/reports/subgraph_{comm_name.lower().replace(" ", "_")}.png')
    plt.close()

draw_subgraph('community_id', 'Louvain')
```
